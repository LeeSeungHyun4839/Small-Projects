```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from collections import Counter
```

```python
def normal(x_i):
    x = np.linspace(x_i-5,x_i+5,100)
    return x, norm.pdf(x-x_i)

def histogram(Max, num):
    x = np.random.randint(0, Max, num)
    bins = np.arange(Max)-0.5
    plt.hist(x, bins)
    plt.show()
    return x

def graph_KDE(List):
    Min, Max = min(List)-5, max(List)+5
    x = np.linspace(Min, Max, 500)
    y = np.zeros((500,))
    for sample in List:
        y += norm.pdf(x-sample)
    return x,y

def f(x, counter):
    x = round(x)
    if x in counter.keys():
        return counter(x)
    else:
        return 0    

def histogram2(Max, num):
    sample = np.random.randint(0, Max, num)
    counter = Counter(x)
```
