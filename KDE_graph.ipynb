```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from collections import Counter
```

```python
def normal(x_i):
    x = np.linspace(x_i-5,x_i+5,100)
    return x, norm.pdf(x-x_i)

def graph_KDE(List, h=1):
    Min, Max = min(List)-5, max(List)+5
    n = len(List)
    x = np.linspace(Min, Max, 500)
    y = np.zeros((500,))
    for sample in List:
        y += norm.pdf((x-sample)/h) / (n*h)
    return x,y

def f(x, counter):
    x = round(x)
    if x in counter.keys():
        return counter[x]
    else:
        return 0    

def histogram2(List):
    Max = max(List)
    Min = min(List)
    counter = Counter(List)
    bins = np.linspace(Min-1, Max+1, 500)
    y = np.zeros((500,))
    for i in range(500):
        y[i] = f(bins[i], counter)
    plt.plot(bins, y, c= 'r')
```

```python
x_i = normal(3)
plt.plot(x_i[0], x_i[1])
```

```python
x = np.random.randint(0, 5, 20)
x = list(x)

print(Counter(x))

histogram2(x)
KDEdata = graph_KDE(x)
plt.plot(KDEdata[0], KDEdata[1]*20, c='b')
plt.show()

plt.plot(KDEdata[0], KDEdata[1], c='b')
plt.show()
```
